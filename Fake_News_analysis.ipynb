```python
nltk.download('wordnet')
lm =  WordNetLemmatizer()
corpus = []
for i in range (len(df)):
    review = re.sub('a-zA-Z0-9',' ', df['title'][i])
    review = review.lower()
    review= review.split()
    review= [lm.lemmatize(x) for x in review if x not in stopwords]
    review= " ".join(review)
    corpus.append(review)
```

```python
len(corpus)
```

## convert text into vector 

```python
tf  = TfidfVectorizer()
x =tf.fit_transform(corpus).toarray()
x
```

```python
y = df['label']
y.head()
```

```python
x_train , x_test , y_train, t_test= train_test_split(x,y, test_size=0.3 ,random_state=10 ,stratify=y)
x_train.shape , y_train.shape , x_test.shape , y_train.shape
```

```python
rf = RandomForestClassifier()
rf.fit(x_train,y_train)
```

```python
y_pred = rf.predict(x_test)
accuracy_score = accuracy_score(t_test, y_pred)
accuracy_score
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

class Evaluation:
    def __init__(self, model, x_train, x_test, y_train, y_test):
        self.model = model
        self.x_train = x_train
        self.x_test = x_test
        self.y_train = y_train
        self.y_test = y_test
    
    def train_evaluation(self):
        y_pred_train = self.model.predict(self.x_train)

        acc_scr_train = accuracy_score(self.y_train, y_pred_train)
        print("Accuracy Score On Training Data Set:", acc_scr_train)
        print()

        con_mat_train = confusion_matrix(self.y_train, y_pred_train)
        print("Confusion Matrix On Training Data Set:\n", con_mat_train)
        print()

        class_rep_train = classification_report(self.y_train, y_pred_train)
        print("Classification Report On Training Data Set:\n", class_rep_train)

    def test_evaluation(self):
        y_pred_test = self.model.predict(self.x_test)

        acc_scr_test = accuracy_score(self.y_test, y_pred_test)
        print("Accuracy Score On Test Data Set:", acc_scr_test)
        print()

        con_mat_test = confusion_matrix(self.y_test, y_pred_test)
        print("Confusion Matrix On Test Data Set:\n", con_mat_test)
        print()

        class_rep_test = classification_report(self.y_test, y_pred_test)
        print("Classification Report On Test Data Set:\n",class_rep_test)
```

```python
Evaluation(rf, x_train,x_test,y_train,t_test).train_evaluation()
```
